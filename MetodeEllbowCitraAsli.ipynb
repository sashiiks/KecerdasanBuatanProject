Menunjukan Grafik metode Ellbow untuk Mata Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2

# Membaca gambar
image = cv2.imread('./FIRE/Images/A01_1.jpg')

# Mengubah gambar menjadi vektor fitur
pixels = image.reshape(-1, 3)  # Mengubah matriks gambar menjadi vektor 1D

# Menentukan jumlah cluster yang akan diuji
clusters = range(1, 10)

# Inisialisasi list untuk menyimpan nilai inersia
inertia_values = []

# Melakukan clustering dengan metode Elbow
for k in clusters:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)
    inertia_values.append(kmeans.inertia_)

# Menampilkan plot Elbow
plt.plot(clusters, inertia_values, 'bo-')
plt.xlabel('Jumlah Clusters')
plt.ylabel('Inersia')
plt.title('Metode Elbow untuk Clustering Gambar')
plt.show()
```

Menunjukan Grafik Dan nilai cluster Untuk Mata Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2

# Membaca gambar
image = cv2.imread('./FIRE/Images/A01_1.jpg')

# Mengubah gambar menjadi format yang dapat diolah
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
pixels = image.reshape(-1, 3)

# Menentukan jumlah cluster yang akan diuji
clusters = range(1, 10)

# Inisialisasi list untuk menyimpan nilai inersia
inertia_values = []

# Melakukan clustering dengan metode Elbow
for k in clusters:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)
    inertia_values.append(kmeans.inertia_)

# Menampilkan plot Elbow
plt.plot(clusters, inertia_values, 'bo-')
plt.xlabel('Jumlah Clusters')
plt.ylabel('Inersia')
plt.title('Metode Elbow untuk Clustering Gambar')
plt.show()

# Menghitung rata-rata nilai dari hasil clustering
best_k = 3  # Jumlah cluster terbaik berdasarkan grafik Elbow
kmeans = KMeans(n_clusters=best_k)
kmeans.fit(pixels)
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_

average_values = []
for i in range(best_k):
    cluster_points = pixels[cluster_labels == i]
    average_values.append(np.mean(cluster_points, axis=0))

print("Rata-rata nilai dari hasil clustering:")
for i, avg_value in enumerate(average_values):
    print(f"Cluster {i+1}: {avg_value}")
```
