Menunjukan Grafik Dari Hasil Segmentasi Mata Normal

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans

# Fungsi untuk menghitung inersia dengan metode Elbow
def calculate_inertia(data, clusters):
    inertia_values = []
    for k in clusters:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        inertia_values.append(kmeans.inertia_)
    return inertia_values

# Membaca gambar
image = cv2.imread('./FIRE/Images/A01_1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Mengambil saluran hijau (green channel)
green_channel = image[:, :, 1]

# Menerapkan CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_output = clahe.apply(green_channel)

# Melakukan Morphological Opening (erosion diikuti dilasi) untuk menghilangkan noise
kernel = np.ones((5, 5), np.uint8)
opened_image = cv2.morphologyEx(clahe_output, cv2.MORPH_OPEN, kernel)

# Menerapkan Matched Filter untuk meningkatkan deteksi
template = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]])
matched_filter_output = cv2.filter2D(opened_image, -1, template)

# Menampilkan gambar hasil tahapan Matched Filter
plt.subplot(1, 2, 1)
plt.imshow(matched_filter_output, cmap='gray')
plt.title('Matched Filter Output')
plt.axis('off')

# Menampilkan plot Elbow untuk tahapan CLAHE
pixels = clahe_output.reshape(-1, 1)
clusters = range(1, 10)
inertia_values = calculate_inertia(pixels, clusters)

plt.subplot(1, 2, 2)
plt.plot(clusters, inertia_values, 'bo-')
plt.xlabel('Jumlah Clusters')
plt.ylabel('Inersia')
plt.title('Metode Elbow untuk CLAHE')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2

# Fungsi untuk menghitung inersia dengan metode Elbow
def calculate_inertia(data, clusters):
    inertia_values = []
    for k in clusters:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        inertia_values.append(kmeans.inertia_)
    return inertia_values

# Membaca gambar
image = cv2.imread('./FIRE/Images/A01_1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Mengambil saluran hijau (green channel)
green_channel = image[:, :, 1]

# Menerapkan CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_output = clahe.apply(green_channel)

# Melakukan Morphological Opening (erosion diikuti dilasi) untuk menghilangkan noise
kernel = np.ones((5, 5), np.uint8)
opened_image = cv2.morphologyEx(clahe_output, cv2.MORPH_OPEN, kernel)

# Menerapkan Matched Filter untuk meningkatkan deteksi
template = np.array([[1, 1, 1], [1, -8, 1], [1, 1, 1]])
matched_filter_output = cv2.filter2D(opened_image, -1, template)

# Menampilkan gambar hasil tahapan Matched Filter
plt.subplot(1, 2, 1)
plt.imshow(matched_filter_output, cmap='gray')
plt.title('Matched Filter Output')
plt.axis('off')

# Menampilkan plot Elbow untuk tahapan CLAHE
pixels = clahe_output.reshape(-1, 1)
clusters = range(1, 10)
inertia_values = calculate_inertia(pixels, clusters)

plt.subplot(1, 2, 2)
plt.plot(clusters, inertia_values, 'bo-')
plt.xlabel('Jumlah Clusters')
plt.ylabel('Inersia')
plt.title('Metode Elbow untuk CLAHE')
plt.tight_layout()
plt.show()

# Menghitung rata-rata nilai dari hasil clustering pada tahapan CLAHE
best_k = 3  # Jumlah cluster terbaik berdasarkan grafik Elbow
kmeans = KMeans(n_clusters=best_k)
kmeans.fit(pixels)
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_

average_values = []
for i in range(best_k):
    cluster_points = pixels[cluster_labels == i]
    average_values.append(np.mean(cluster_points))

print("Rata-rata nilai dari hasil clustering:")
for i, avg_value in enumerate(average_values):
    print(f"Cluster {i+1}: {avg_value}")
```
